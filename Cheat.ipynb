```python

import nltk
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

# Function to map POS tags to WordNet tags
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  # Default to noun if no match

def lemmatize_sentence(sentence):
    # Tokenize the sentence
    tokens = nltk.word_tokenize(sentence)
    # Get the POS tags for each token
    tagged_tokens = nltk.pos_tag(tokens)
    lemmatizer = WordNetLemmatizer()
    lemmatized_sentence = []
    for token, tag in tagged_tokens:
        # Map NLTK POS tags to WordNet POS tags
        wordnet_pos = get_wordnet_pos(tag)
        # Lemmatize the token
        lemmatized_token = lemmatizer.lemmatize(token, pos=wordnet_pos)
        lemmatized_sentence.append(lemmatized_token)
    return ' '.join(lemmatized_sentence)

# Example usage
sentence = "The striped bats are hanging on their feet for best"
lemmatized = lemmatize_sentence(sentence)
print(lemmatized)
```

```python

# 19/07/24
# similarity between two different sentence and word
import nltk
nltk.download('wordnet')
nltk.download('punkt')
import spacy
from nltk.corpus import wordnet

nlp= spacy.load("en_core_web_sm")


def word_similarity(word1,word2):
    """
    Calculate the similarity between two word using wordnet from NLTK.
    """
    sysnet1=wordnet.synsets(word1)[0]
    sysnet2=wordnet.synsets(word2)[0]
    similarity=sysnet1.wup_similarity(sysnet2)
    return similarity


def sentence_similarity(sentence1,sentence2):
    """
    Calculate the similarity between two sentence using spaCy's pre-trained model.
    """

    doc1=nlp(sentence1)
    doc2=nlp(sentence2)
    similarity=doc1.similarity(doc2)
    return similarity

word1="sparrow"
word2="cat"
word_sim=word_similarity(word1,word2)
print(f"Similarity between word'{word1}' and '{word2}': {word_sim:4f}")

sentence1="The quick brown fox jumps over the lazy dog."
sentence2="A fast dark fox leap over a sleepy dog."
sentence_sim=sentence_similarity(sentence1,sentence2)
print(f"Similarity between sentence'{sentence1}' and '{sentence2}': {sentence_sim:4f}")
```

fuzz.ratio: Measures the similarity between two strings by calculating the Levenshtein distance.

process.extractOne: Finds the closest match in a list of strings.

process.extractBests: Finds all matches that meet a specified similarity threshold.

fuzz.partial ratio: Measures similarity by considering the best matching substring, which is useful when one sting is subset of another.


```python
import pandas as pd
from thefuzz import process, fuzz


data = {
    "Company Name": [
        "Google Inc.", "Google LLC", "Goooogle", "GooGle!", "Apple Inc.", "Apple", "Microsoft Corp.",
        "Microsoft Corporation", "MicroSoft", "Amazon.com", "Amazon Inc", "Amaaaazon"
    ]
}


df = pd.DataFrame(data)

def standardize_company_names(df, threshold=80):

    # Create a dictionary to map original names to standardized names
    standard_name_map = {}

    # Iterate over each company name
    for name in df['Company Name']:
        if name not in standard_name_map:
            # Find the best match for the current name within existing names in the map
            match_result = process.extractOne(name, standard_name_map.keys(), scorer=fuzz.partial_ratio, score_cutoff=threshold)

            if match_result:
                match, score = match_result
                standard_name_map[name] = standard_name_map[match]
                #print(standard_name_map[match])
            else:
                standard_name_map[name] = name


    # Replace the names in the DataFrame with the standardized names
    df['Standardized Company Name'] = df['Company Name'].map(standard_name_map)

    return df

df_standardized = standardize_company_names(df)
print("\nStandardized Data:\n", df_standardized)
```
